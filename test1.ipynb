```python
import ollama
```

```python
# GENERATION
response = ollama.generate(model="tinyllama:latest",prompt="hello!")
response.response
```

```python
res = ollama.generate(model="tinyllama:latest",prompt="hello")
res.response
```

```python
# STREAM RESPONSE
response = ollama.generate(model="tinyllama:latest",prompt="why does moon glow ?", stream=True)
for i in response:
    print(i["response"], end="")
```

```python
# SYSTEM INSTRUCTIONS
response = ollama.generate(model="tinyllama:latest",prompt="why does moon glow ?", system="You are an funny assistant , you explain things in funny way")
print(response.response)
```

```python
# text wrap for lining 
import ollama
import textwrap

response = ollama.generate(
    model="tinyllama:latest",
    prompt=" moon ",
    system="You are a poet assistant. Explain things in a poetry way."
)

print(textwrap.fill(response.response, width=100))
```

```python
# OPTIONS PARAMETER
response = ollama.generate(model="tinyllama:latest",prompt="why is the ocean blue",
                           options={
                               "temperature":0.3,
                               "top_p":0.5,
                               "top_k":45
                           })


print(textwrap.fill(response.response, width=100))
```
